```
data10=data4.groupby(['author','product'])['score'].max().unstack() 
data10
```

```
data11=data10.dropna(thresh=50,axis=1).fillna(0)
data11
```

```
mobile_similarity_df=data11.corr(method='pearson')#inbuilt method
mobile_similarity_df.head(3)
```

```
def recommend_mobile(mobile_name,score_mobile):
  similar_score=mobile_similarity_df[mobile_name]*(score_mobile-2.5)
  similar_score=similar_score.sort_values(ascending=False)
  return similar_score
```

```
mobile_name=[("Samsung Galaxy Ace 2",5),("Sony Xperia miro",1),("Apple iPhone 5 32GB",5)]
similar_mobile=pd.DataFrame()
for mobile,rating in mobile_name:
  similar_mobile=similar_mobile.append(recommend_mobile(mobile,rating),ignore_index=True)
similar_mobile.head()
similar_mobile.sum().sort_values(ascending=False)
```
